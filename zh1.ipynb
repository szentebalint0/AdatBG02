```sql
--Feladat1
SELECT szall.TIPUS, ROUND(COUNT(*), -2) AS darabszám
FROM Foglalas f JOIN Szoba sz ON f.SZOBA_FK  = sz.SZOBA_ID JOIN Szallashely szall ON sz.SZALLAS_FK = szall.SZALLAS_ID
GROUP BY szall.TIPUS
HAVING ROUND(COUNT(*), -2) < 500
```

```sql
--feladat 2
SELECT szall.SZALLAS_NEV, sz.SZOBA_SZAMA, COUNT(*) AS 'Foglalások száma'
FROM Szoba sz JOIN Szallashely szall ON sz.SZALLAS_FK = szall.SZALLAS_ID LEFT JOIN Foglalas f ON sz.SZOBA_ID = f.SZOBA_FK
WHERE f.SZOBA_FK IS NOT NULL
GROUP BY szall.SZALLAS_NEV, sz.SZOBA_SZAMA
```
